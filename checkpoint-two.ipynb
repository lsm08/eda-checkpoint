```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
df = pd.read_csv('./ds_salaries.csv') 
df.head()# displays the first 5 rows of a dataframe

```

## Get to Know the Numbers

Now that you have everything setup, put any code that you use to get to know the dataframe and its rows and columns better in the cell below. You can use whatever techniques you like, except for visualizations. You will put those in a separate section.

```python
# df.describe # to get stastics for the numerical columns
print(df.describe)
# shows the summary of the data
print(df.info)  

# displays all column names
print(df.columns) 

missing_values = df.isnull().sum() # finding missing values in data
print(missing_values) 

df.nunique() # displays unique values in each column of the data

duplicate_rows = df[df.duplicated(keep='first')]
num_duplicates = duplicate_rows.shape[0]
print(num_duplicates)

```

```python
average_salary = df[['job_title','salary_in_usd']].groupby('job_title').mean().sort_values(by='salary_in_usd',ascending = False)
print(average_salary)
top5_high_salaries = average_salary.head(5)
print(top5_high_salaries)
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# Creating a plot to show top 5 jobs by salary
plt.figure(figsize=(20, 6))
print(top5_high_salaries)
plt.bar(top5_high_salaries.index, top5_high_salaries ['salary_in_usd'])
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.title('Top 5 Highest Paid Data Science Jobs')
plt.show()
```

```python
#Visualize salaries by company location
# print('columns', df.columns)
salary_by_location = df[['company_location','salary_in_usd']].groupby('company_location').mean().sort_values(by='salary_in_usd', ascending = False)
plt.figure(figsize=(20, 5))

plt.bar(salary_by_location.index, salary_by_location['salary_in_usd'], )
plt.title('Average Salary by Location')
plt.xlabel('Company Location')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.show()
```
